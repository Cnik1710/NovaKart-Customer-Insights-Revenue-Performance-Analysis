### -----------------------------------------------------------------------------------------------------------------------------

### Feature Engineering

```python
df_combined = df_combined.rename(columns={'customer_num': 'customer_id', 'paid': 'revenue', 'product_num':'product_id', 'amount':'product_qty', 'cost':'product_cost'})
```

```python
df_combined['year'] = df_combined['purch_date'].dt.year
```

```python
df_combined['month'] = df_combined['purch_date'].dt.month_name()
```

```python
df_combined['day'] = df_combined['purch_date'].dt.day_name()
```

```python
df_combined['profit'] = df_combined['revenue'] - df_combined['product_cost']
```

```python
df_combined['profit_margin'] = np.where(df_combined['revenue'] != 0, (df_combined['profit'] / df_combined['revenue']) * 100, 0)
```

```python
cols_order = ['customer_id', 'full_name', 'gender', 'city', 'state', 'product_id', 'product', 'company', 
              'product_qty', 'product_cost', 'purch_date', 'year', 'month', 'day', 'revenue', 'profit', 'profit_margin']

df_combined = df_combined[cols_order]

df_combined.head()
```

### 1. Customer & Location Insights

### (a) Customer Distribution by Gender

```python
gender_counts = (df_combined[['customer_id', 'gender']].drop_duplicates().groupby('gender').size())

custom_colors = ['#1f77b4', '#ff7f0e', '#db3939'] 

plt.figure(figsize=(5,5))

plt.pie(gender_counts, labels=gender_counts.index, autopct="%1.1f%%", colors=custom_colors) 

plt.title("Customer Distribution by Gender")
plt.show()
```

### (b) Customer Distribution by States

```python
df_states = df_combined[df_combined['state'] != 'N/A']

state_counts = df_states.groupby('state')['customer_id'].nunique().sort_values(ascending=False).head(5)

state_counts.iloc[::-1].plot(kind="barh", figsize=(10,5))

plt.title("Customer Distribution by States")
plt.xlabel('Customer Count')
plt.ylabel('States')
plt.show()
```

### (c) Customer Distribution by Cities

```python
df_cities = df_combined[df_combined['city'] != 'N/A']

city_counts = df_cities.groupby('city')['customer_id'].nunique().sort_values(ascending=False).head(5)

city_counts.iloc[::-1].plot(kind="barh", figsize=(10,5))

plt.title("Customer Distribution by Cities")
plt.xlabel('Customer Count')
plt.ylabel('Cities')
plt.show()
```

### (d) Revenue & Profit Generation by Gender

```python
df_gender = df_combined[df_combined["gender"] != "N/A"] 

rp_gender = df_gender.groupby("gender")[["revenue","profit"]].sum().reindex(["Male", "Female"])

rp_gender.plot(kind="bar", figsize=(8,5)) 
plt.title("Revenue & Profit Generation by Gender")
plt.xlabel("Gender")
plt.ylabel("Amount")
plt.xticks(rotation=0)
plt.legend()
plt.show()
```

### (e) Revenue & Profit Generation by Cities

```python
df_city = df_combined[df_combined["city"] != "N/A"]

city_rev = df_city.groupby("city")[["revenue","profit"]].sum().sort_values("revenue", ascending=False).head(5)

city_rev.plot(kind="bar", figsize=(12,6))
plt.title("Revenue & Profit Generation by Cities")
plt.xlabel("Cities")
plt.ylabel("Revenue & Profit")
plt.xticks(rotation=0)
plt.legend()
plt.show()
```

### (f) Revenue & Profit Generation by States

```python
df_state = df_combined[df_combined["state"] != "N/A"]

state_rev = df_state.groupby("state")[["revenue","profit"]].sum().sort_values("revenue", ascending=False).head(5)

state_rev.plot(kind="bar", figsize=(12,6))
plt.title("Revenue & Profit Generation by States")
plt.xlabel("States")
plt.ylabel("Revenue & Profit")
plt.xticks(rotation=0)
plt.legend()
plt.show()
```

### (g) Revenue & Profit Generation by Customers

```python
customer_rev = df_combined.groupby("full_name")[["revenue","profit"]].sum().sort_values("revenue", ascending=False).head(5)

customer_rev.plot(kind="bar", figsize=(12,5))
plt.title("Revenue & Profit Generation by Customers")
plt.xlabel("Customers")
plt.ylabel("Revenue & Profit")
plt.xticks(rotation=0)
plt.legend()
plt.show()
```

### (h) Average Profit Margin (%) by States

```python
df_state = df_combined[df_combined['state'] != 'N/A']

pm_state = df_state.groupby("state")["profit_margin"].mean().sort_values(ascending=False).head(5)

pm_state.plot(kind="bar", figsize=(12,5))
plt.title("Average Profit Margin (%) by States")
plt.xlabel("States")
plt.ylabel("Average Profit Margin (%)")
plt.xticks(rotation=0)
plt.show()
```

### (i) Average Profit Margin (%) by Cities

```python
df_city = df_combined[df_combined['city'] != 'N/A']

pm_city = df_city.groupby("city")["profit_margin"].mean().sort_values(ascending=False).head(5)

pm_city.plot(kind="bar", figsize=(12,5))
plt.title("Average Profit Margin (%) by Cities")
plt.xlabel("Cities")
plt.ylabel("Average Profit Margin (%)")
plt.xticks(rotation=0)
plt.show()
```

### 2. Product & Company Insights

### (a) Revenue & Profit Generation by Months

```python
rev_month = df_combined.groupby("month")[["revenue", "profit"]].sum()

month_order = ["January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December"]

rev_month = rev_month.reindex(month_order).fillna(0)

ax = rev_month.plot(kind="line", marker="o", figsize=(15,5))
plt.title("Revenue & Profit Generation by Months")
plt.xlabel("Months")
plt.ylabel("Revenue & Profit")
plt.xticks(ticks=range(len(month_order)), labels=month_order, rotation=0)
plt.legend()
plt.show()
```

### (b) Revenue & Profit Generation by Weekdays

```python
rev_day = df_combined.groupby("day")[["revenue","profit"]].sum()

dayname_order = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]

rev_dayname = rev_day.reindex(dayname_order).fillna(0)

ax = rev_dayname.plot(kind="line", marker="o", figsize=(15,5))
plt.title("Revenue & Profit Generation by Weekdays")
plt.xlabel("Weekdays")
plt.ylabel("Revenue & Profit")
plt.xticks(ticks=range(len(dayname_order)), labels=dayname_order, rotation=0)
plt.legend()
plt.show()
```
